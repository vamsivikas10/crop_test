```python
# overview of  repeated , distinct and null values in each column
def check_repeated_distinct_value_columns_check_null_values(df):
    count_of_repeatedvalues=0
    for everycolumn in df.columns :
        for i in range(total_rows(df)):
            first_value =df[everycolumn].loc[0]
            if first_value == df[everycolumn].loc[i]:
                count_of_repeatedvalues+=1    
        if count_of_repeatedvalues == total_rows(df):
           print("repeated values in :",everycolumn)
            
        
        print(everycolumn , f" has  {df[everycolumn].nunique()}  distinct values")

        is_null = df[everycolumn].isnull().any()
        if  is_null == True:
            print(f"{everycolumn} has {df[everycolumn].isnull().sum()} null values")

        print(" ")

#no of rows and columns in dataframe
def total_rows(df):
    total_rows = df.shape[0]
    return total_rows
def total_columns(df):  
    total_columns = df.shape[1]
    return total_columns

```

```python
def verify_relationship(df):
    # Only check rows where all three values exist and are positive
    mask = (df['Area harvested'].notnull() & 5
            df['Yield'].notnull() & 
            df['Production'].notnull() &
            (df['Area harvested'] > 0) & 
            (df['Yield'] > 0))
    
    if mask.sum() == 0:
        return "No valid rows to verify relationship"
    
    calculated_production = df.loc[mask, 'Area harvested'] * df.loc[mask, 'Yield']
    actual_production = df.loc[mask, 'Production']
    
    # Calculate absolute and relative differences
    abs_diff = (calculated_production - actual_production).abs()
    rel_diff = abs_diff / actual_production
    
    print(f"Verification for {mask.sum()} valid rows:")
    print(f"Absolute difference stats:\n{abs_diff.describe()}")
    print(f"\nRelative difference stats:\n{rel_diff.describe()}")
    
    # Check what percentage are within 1% tolerance
    within_tolerance = (rel_diff <= 0.01).mean()
    print(f"\n{within_tolerance*100:.2f}% of rows within 1% tolerance")

verify_relationship(Final_Crop_df)
```
